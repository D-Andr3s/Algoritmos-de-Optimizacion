## Introduci√≥n

En este trabajo, abordaremos un problema donde emplearemos diferentes tipos de algoritmos para lograr una mejor optimizaci√≥n. En la industria cinematogr√°fica, organizar eficientemente las sesiones de doblaje es un desaf√≠o cr√≠tico, ya que cada toma requiere la presencia de actores espec√≠ficos, y los costos se disparan si estos deben asistir repetidamente a m√∫ltiples d√≠as de grabaci√≥n. Aunque aparentemente sencillo, este problema esconde una complejidad combinatoria que lo sit√∫a en la clase de los problemas NP-duros.

Para resolverlo, utilizamos dos enfoques complementarios:

* Un algoritmo exacto (programaci√≥n lineal), que garantiza la soluci√≥n √≥ptima mediante modelos matem√°ticos.

* Un algoritmo heur√≠stico (algoritmos gen√©ticos), que busca soluciones cercanas al √≥ptimo en tiempos reducidos.

Con esto, ponemos a prueba ambas herramientas, validando un principio fundamental en optimizaci√≥n: "No existe un algoritmo √∫nico que resuelva todos los problemas, sino que se requiere elegir la t√©cnica m√°s adecuada seg√∫n el contexto".

#Modelo

Para abordar este problema, emplearemos un algoritmo exacto basado en t√©cnicas de programaci√≥n lineal. A continuaci√≥n, desglosamos los componentes del modelo que nos permitir√° resolverlo.

1. Representaci√≥n del Espacio de Soluciones

El espacio de soluciones comprende todas las posibles formas v√°lidas de asignar 30 tomas a 5 d√≠as, respetando las restricciones impuestas. Utilizamos variables binarias para esta representaci√≥n, las cuales indican si una toma espec√≠fica se asigna a un d√≠a determinado y si un actor debe estar presente en ese d√≠a.

* Variables asignaci√≥n de tomas:
 $$ x_{td} = \begin{cases} 1 & \text{si la toma } t \text{ se asigna al d√≠a } d \\ 0 & \text{en otro caso} \end{cases} $$

Donde $ùë°‚àà{1,2,‚Ä¶,30}$ y $ùëë‚àà{1,2,‚Ä¶,5}$.

ejemplo

Si $x_{5,2}=1$, si la toma 5 se grabara el dia 2.

Si $x_{10,3}=0$, si la toma 10 no se grabara el dia 3.

* Variables de presencia de actores:
 $$ a_{ad} = \begin{cases} 1 & \text{si el actor } a \text{ debe asistir al d√≠a } d \\ 0 & \text{en otro caso} \end{cases} $$

Donde $ùëé‚àà{1,2,‚Ä¶,10}$.

ejemplo

Si $a_{3,1}=1$, el actor 3 debe ir al estudio el dia 1.

Si $a_{7,4}=0$, el actor 7 no trabaja el dia 4.


2. Funci√≥n Objetivo

El objetivo es minimizar el gasto total, que se traduce en reducir la cantidad de d√≠as que cada actor debe asistir al estudio. La funci√≥n objetivo es:

$$ \text{Minimizar } \sum_{a=1}^{10} \sum_{d=1}^{5} a_{ad} $$

Explicaci√≥n: Si un actor a asiste 3 d√≠as, contribuye con 3 unidades al costo total. La funci√≥n suma los d√≠as de trabajo de todos los actores y busca reducir esta suma al m√≠nimo posible.

3. Implementaci√≥n de las Restricciones

* Asignaci√≥n √∫nica de tomas: Cada toma debe asignarse a exactamente un d√≠a. No se permite grabar una toma en m√∫ltiples d√≠as.
 $$ \sum_{d=1}^{5} x_{td} = 1 \quad \forall \, t \in \{1, 2, \ldots, 30\} $$

* L√≠mite de tomas por d√≠a: No se pueden grabar m√°s de 6 tomas por d√≠a. Esto impone un equilibrio en la carga de trabajo diaria.
$$ \sum_{t=1}^{30} x_{td} \leq 6 \quad \forall \, d \in \{1, 2, \ldots, 5\} $$

* Relaci√≥n entre actores y d√≠as: Si una toma t requiere al actor a, entonces el actor a debe asistir al d√≠a d donde se grabe t. Esto evita que un actor falte cuando se graba una toma que lo necesita.
$$ a_{ad} \geq x_{td} \quad \forall \, d, \, \forall \, t \text{ donde el actor } a \text{ participa en la toma } t $$.

A continuaci√≥n, se proceder√° a probar dos m√©todos de algoritmos de optimizaci√≥n con el objetivo de comparar las soluciones √≥ptimas obtenidas. Para finalizar esta etapa y avanzar hacia la implementaci√≥n del c√≥digo, definiremos los pasos necesarios para establecer la conexi√≥n entre el modelo matem√°tico y la soluci√≥n computacional

```
# --------------------------------------------
# Algoritmo exacto-Programacion Lineal (PL)
# --------------------------------------------
# --------------------------------------------
# Cargar librer√≠as
# --------------------------------------------
import gdown
import pandas as pd
import pulp as lp

# --------------------------------------------
# Cargar base de datos
# --------------------------------------------

# Descargar el archivo
file_id = "1VCnAJ5KJk5kliM92EBhj-egRuPo20ejZ"
output_file = "Datos_problema.xlsx"
gdown.download(f"https://drive.google.com/uc?id={file_id}", output_file, quiet=False)

# Leer el archivo ignorando filas no deseadas
df = pd.read_excel(
    output_file,
    engine='openpyxl',
    header=None,
    skiprows=[0, 1]  # Ignorar fila 0 (metadatos) y fila 1 (encabezados)
)

# Eliminar columnas vac√≠as y asignar nombres
df = df.dropna(axis=1, how='all')
columnas = ["Toma", "Actor1", "Actor2", "Actor3", "Actor4",
            "Actor5", "Actor6", "Actor7", "Actor8", "Actor9", "Actor10", "Total"]
df.columns = columnas

# Eliminar filas que contienen 'TOTAL' o est√°n vac√≠as
df = df[~df['Toma'].astype(str).str.contains('TOTAL|nan', na=False)]

# Convertir a enteros y establecer "Toma" como √≠ndice
df = df.fillna(0)
df['Toma'] = df['Toma'].astype(int)
for col in df.columns.drop('Toma'):
    df[col] = df[col].astype(int)
df.set_index("Toma", inplace=True)

# Mostrar DataFrame
print("¬°Datos listos para el modelo!")
display(df.head())

# =============================================================================
# CONFIGURACI√ìN INICIAL: DATOS Y PAR√ÅMETROS
# =============================================================================
tomas = df.index.tolist()
actores = df.columns.drop("Total").tolist()
dias_max = 5  # 5 d√≠as
dias = list(range(1, dias_max + 1))

# =============================================================================
# PASO 1: DEFINIR EL PROBLEMA DE OPTIMIZACI√ìN
# =============================================================================
modelo = lp.LpProblem(name="Programacion_Tomas", sense=lp.LpMinimize)

# =============================================================================
# PASO 2: DEFINIR VARIABLES DE DECISI√ìN
# =============================================================================
x = lp.LpVariable.dicts(
    "TomaDia",
    [(t, d) for t in tomas for d in dias],
    cat=lp.LpBinary
)

y = lp.LpVariable.dicts(
    "ActorDia",
    [(a, d) for a in actores for d in dias],
    cat=lp.LpBinary
)

# =============================================================================
# PASO 3: AGREGAR LA FUNCI√ìN OBJETIVO
# =============================================================================
modelo += lp.lpSum(y[a, d] for a in actores for d in dias), "Costo_Total_Actores"

# =============================================================================
# PASO 4: AGREGAR RESTRICCIONES
# =============================================================================

# Restricci√≥n 1: Cada toma asignada a exactamente un d√≠a
for t in tomas:
    modelo += (
        lp.lpSum(x[t, d] for d in dias) == 1,
        f"Asignacion_Unica_Toma_{t}"
    )

# Restricci√≥n 2: Exactamente 6 tomas por d√≠a
for d in dias:
    modelo += (
        lp.lpSum(x[t, d] for t in tomas) == 6,
        f"Exactamente6Tomas_Dia_{d}"
    )

# Restricci√≥n 3: Relaci√≥n toma-actor-d√≠a
for a in actores:
    tomas_con_actor = df[df[a] == 1].index.tolist()
    for t in tomas_con_actor:
        for d in dias:
            modelo += (
                y[a, d] >= x[t, d],
                f"Asistencia_{a}_Toma{t}_Dia{d}"
            )

# =============================================================================
# PASO 5: RESOLVER EL MODELO
# =============================================================================
try:
    resultado = modelo.solve(
        solver=lp.PULP_CBC_CMD(
            msg=True,
            timeLimit=300  # Puedes aumentar el tiempo l√≠mite si es necesario
        )
    )
except lp.PulpSolverError as e:
    print("Error en el solver:", e)
    print("Revisa la instalaci√≥n de CBC o prueba con otro solver.")

# =============================================================================
# PASO 6: INTERPRETAR RESULTADOS
# =============================================================================
if lp.LpStatus[modelo.status] == "Optimal":
    # Calcular d√≠as usados
    dias_usados = []
    for d in dias:
        if sum(x[t, d].varValue for t in tomas) > 0:
            dias_usados.append(d)

    # Mostrar m√©tricas clave
    print(f"\n¬°Soluci√≥n √≥ptima encontrada!")
    print(f"D√≠as requeridos: {len(dias_usados)}")
    print(f"Total d√≠as-actores: {int(lp.value(modelo.objective))}")

    # Mostrar asignaci√≥n de tomas por d√≠a
    for d in dias_usados:
        tomas_dia = [t for t in tomas if x[t, d].varValue > 0.9]
        print(f"\nD√≠a {d} ({len(tomas_dia)} tomas):")
        print(f"Tomas: {tomas_dia}")

    # Mostrar d√≠as de asistencia por actor
    print("\nAsistencia de actores:")
    for a in actores:
        dias_asistencia = [d for d in dias if y[a, d].varValue > 0.9]
        print(f"- {a}: {len(dias_asistencia)} d√≠as ({dias_asistencia})")
else:
    print("No se encontr√≥ soluci√≥n √≥ptima. Estado:", lp.LpStatus[modelo.status])

```

#An√°lisis

### ¬øQue complejidad tiene el problema?

El problema de organizar las sesiones de doblaje es NP-duro, perteneciente a una clase de problemas donde no se conocen algoritmos exactos eficientes para instancias grandes. Su complejidad surge de dos factores clave:

* Restricciones combinatorias: Asignar 30 tomas a 5 d√≠as (6 tomas/d√≠a) mientras se minimiza la superposici√≥n de actores.

* Crecimiento factorial del espacio de soluciones: El n√∫mero de combinaciones v√°lidas crece exponencialmente con el tama√±o de entrada, vinculando el  problemas de Bin Packing (empaquetamiento en contenedores) y Set Cover (cobertura de conjuntos) son dos desaf√≠os cl√°sicos en optimizaci√≥n combinatoria.


### An√°lisis de Complejidad y Espacio de Soluciones  
  
El problema de organizar sesiones de doblaje es un desaf√≠o combinatorio donde se deben asignar 30 tomas a 5 d√≠as (6 tomas/d√≠a), minimizando los d√≠as que los actores deben asistir. Su complejidad surge no solo de las restricciones, sino del enorme espacio de soluciones, como se detalla a continuaci√≥n.

#### 1. Orden de Complejidad

**Programaci√≥n Lineal (LP):**

* Tiempo: $O(2^n)$  en el peor caso, donde $n$ es el n√∫mero de variables binarias (tomas √ó d√≠as + actores √ó d√≠as).

* Espacio: $O(n*m)$, donde $n=30$ tomas y $m=10$ actores, debido a la matriz de restricciones actor-toma-d√≠a.

**Algoritmo Gen√©tico (GA)**:  
- **Tiempo**: (O(G * P * E)), donde:  
  - \(G\): Generaciones (ej: 100),  
  - \(P\): Poblaci√≥n (ej: 300),  
  - \(E\): Coste por evaluaci√≥n (\(O(5 \times 6 \times 10) = O(300)\)).  
- **Ejemplo**: $ 100 \times 300 \times 300 = 9 \times 10^6 $ operaciones.  
- **Espacio**: O(P * n), donde (P = 300) y (n = 30), resultando en 9000 unidades de memoria.

Dada la complejidad computacional, es cr√≠tico entender el espacio de soluciones para justificar el uso de m√©todos heur√≠sticos. Este espacio, que representa todas las posibles asignaciones v√°lidas de tomas a d√≠as, crece factorialmente con el n√∫mero de tomas.
#### 2. Espacio de Soluciones

El n√∫mero total de posibles asignaciones de tomas a d√≠as est√° dado por:

$$\text{Espacio de Soluciones} = \frac{30!}{(6!)^5 \cdot 5!} \approx 1.11 \times 10^{23}$$

**Explicaci√≥n:**

- **$30!$**: Permutaciones de todas las tomas.
- **$(6!)^5$**: Se descarta el orden dentro de cada d√≠a (6 tomas/d√≠a).
- **$5!$**: Se descarta el orden de los d√≠as (los d√≠as no est√°n etiquetados).

Para dimensionar $(10^{23})$, imagina evaluar 1 mill√≥n de soluciones/segundo: tomar√≠a $(3.5 \times 10^9)$ a√±os, superando la edad del universo $((1.38 \times 10^9) a√±os)$. Esto justifica el uso de heur√≠sticas como GA o LP.

# Dise√±o
El problema de organizar las sesiones de doblaje se enmarca dentro de la clase de problemas NP-duros, es decir, aquellos que, aunque pueden verificarse en tiempo polinomial, no admiten soluciones exactas eficientes para instancias grandes debido a su complejidad combinatoria exponencial. Este desaf√≠o, relacionado con variantes del Bin Packing (empaquetamiento de tomas en d√≠as) y el Set Cover Problem (minimizar la superposici√≥n de actores), requiere t√©cnicas especializadas que equilibren la precisi√≥n de la soluci√≥n con la viabilidad computacional.

Para abordarlo, se implementaron dos enfoques complementarios: Programaci√≥n Lineal Entera (PLI) y Algoritmos Gen√©ticos (AG). La elecci√≥n de estas t√©cnicas se fundamenta en sus ventajas distintivas ante problemas combinatorios. La PLI, un m√©todo exacto, permite modelar matem√°ticamente el problema mediante variables binarias que representan la asignaci√≥n de tomas a d√≠as y la participaci√≥n de actores, garantizando la optimalidad mediante solvers como CPLEX o Gurobi. Estos optimizadores aplican t√©cnicas avanzadas (branch-and-bound, cortes) para reducir el espacio de b√∫squeda, lo que la hace ideal para instancias peque√±as (30 tomas, 10 actores) donde se prioriza la precisi√≥n.

Por otro lado, los Algoritmos Gen√©ticos, una heur√≠stica bioinspirada, exploran el espacio de soluciones mediante operadores de selecci√≥n, cruce y mutaci√≥n, imitando la evoluci√≥n natural. Esta t√©cnica no garantiza optimalidad, pero es altamente eficaz en problemas con espacios de soluciones astron√≥micos ($10^{23}$  combinaciones), donde m√©todos exactos son inviables. Su fortaleza radica en la escalabilidad: para casos con cientos de tomas, los AG pueden encontrar soluciones near-optimales en segundos, adapt√°ndose din√°micamente a restricciones cambiantes.

La decisi√≥n de emplear ambos m√©todos se asemeja a la de un fontanero que elige entre una llave precisa y un desatascador seg√∫n la obstrucci√≥n. La PLI act√∫a como la "llave exacta", capaz de ajustar perfectamente las tomas en d√≠as para minimizar costos (resultando en 27 d√≠as-actores), mientras que los AG funcionan como el "desatascador adaptable", resolviendo r√°pidamente configuraciones complejas con un costo cercano al √≥ptimo (28 d√≠as-actores). Esta dualidad no solo valida la soluci√≥n √≥ptima mediante contrastaci√≥n, sino que tambi√©n prepara el terreno para escalar el problema: la PLI para instancias peque√±as y los AG para futuras ampliaciones.

En s√≠ntesis, la combinaci√≥n de t√©cnicas refleja un dise√±o estrat√©gico: aprovechar la precisi√≥n de la PLI para obtener un resultado √≥ptimo verificable y utilizar los AG como alternativa flexible, asegurando que el enfoque se adapte tanto a la complejidad te√≥rica del problema como a sus posibles extensiones pr√°cticas.
