```python
#N Reinas - Vuelta Atrs()
################################################################################

#Verifica que en la solucin parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la solucin tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " est " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la solucin al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(5,solucion=[],etapa=0)
```

### Primer Intento: Fuerza Bruta

podemos idealizar una solucion por fuerza bruta para este problema de encontrar los dos puntos mas cercanos en un cojunto de puntos, este enfoque es sencillo y directo, pero no es eficiente para conjuntos de datos grandes debido a su complejidad temporal de $ O(n^2) $ donde
$ n $ es el nmero de puntos. Esto se debe a que se comparan todas las parejas posibles de puntos.

- La idea principal de la fuerza bruta es comparar todas las parejas posibles de puntos y calcular la distancia entre ellos. Luego, seleccionamos la pareja con la distancia ms pequea.

- Este mtodo es adecuado para conjuntos de datos pequeos o medianos. Para conjuntos de datos grandes, es recomendable utilizar algoritmos ms eficientes, como el de divide y vencers, que tiene una complejidad de $ O(n.log(n)) $.


```python
# -----Dos puntos mas cercanos----
# Dado un conjunto de puntos se trata de encontrar los dos puntos mas cercanos

# Librerias
import random
import math

# Suponemos las listas de numeros
LISTA_1D = [random.randrange(1, 10000) for x in range(1000)]

# Funcin para calcular la distancia entre dos puntos en 1D
def distancia_1d(p1, p2):
    return abs(p1 - p2)

# Funcin para encontrar los dos puntos ms cercanos en 1D (fuerza bruta)
def fuerza_bruta_1d(puntos):
    n = len(puntos)
    if n < 2:
        return None, None, float('inf')  # No hay suficientes puntos

    # Inicializar los puntos ms cercanos y la distancia mnima
    punto1, punto2 = puntos[0], puntos[1]
    min_distancia = distancia_1d(punto1, punto2)

    # Comparar todas las parejas de puntos
    for i in range(n):
        for j in range(i + 1, n):
            dist = distancia_1d(puntos[i], puntos[j])
            if dist < min_distancia:
                min_distancia = dist
                punto1, punto2 = puntos[i], puntos[j]

    return punto1, punto2, min_distancia

# Ejecutar la funcin de fuerza bruta en 1D
p1, p2, dist = fuerza_bruta_1d(LISTA_1D)

# Mostrar resultados
print("En 1D:")
print(f"Los puntos ms cercanos son {p1} y {p2} con una distancia de {dist}")
```

### Segundo Intento: Divide y Vencers

Ahora vamos a implementar el segundo intento utilizando el mtodo de Divide y Vencers para resolver el problema de encontrar los dos puntos ms cercanos en un conjunto de puntos en 1D. Este mtodo es ms eficiente que la fuerza bruta, con una complejidad de $ O(n.logn)$.


```python
# -----Dos puntos mas cercanos----
# Dado un conjunto de puntos se trata de encontrar los dos puntos mas cercanos

# Librerias
import random
import math

# Suponemos las listas de numeros
LISTA_1D = [random.randrange(1, 10000) for x in range(1000)]

# Funcin para calcular la distancia entre dos puntos en 1D
def distancia_1d(p1, p2):
    return abs(p1 - p2)

# Funcin para encontrar los dos puntos ms cercanos en 1D (fuerza bruta)
def fuerza_bruta_1d(puntos):
    n = len(puntos)
    if n < 2:
        return None, None, float('inf')  # No hay suficientes puntos

    # Inicializar los puntos ms cercanos y la distancia mnima
    punto1, punto2 = puntos[0], puntos[1]
    min_distancia = distancia_1d(punto1, punto2)

    # Comparar todas las parejas de puntos
    for i in range(n):
        for j in range(i + 1, n):
            dist = distancia_1d(puntos[i], puntos[j])
            if dist < min_distancia:
                min_distancia = dist
                punto1, punto2 = puntos[i], puntos[j]

    return punto1, punto2, min_distancia

# Funcin para encontrar los dos puntos ms cercanos en 1D (divide y vencers)
def divide_y_venceras_1d(puntos):
    n = len(puntos)
    if n <= 3:
        # Si hay pocos puntos, usar fuerza bruta
        return fuerza_bruta_1d(puntos)

    # Ordenar los puntos
    puntos_ordenados = sorted(puntos)

    # Dividir el conjunto en dos mitades
    mitad = n // 2
    punto_medio = puntos_ordenados[mitad]

    izquierda = puntos_ordenados[:mitad]
    derecha = puntos_ordenados[mitad:]

    # Resolver recursivamente en cada mitad
    p1_izq, p2_izq, d_izq = divide_y_venceras_1d(izquierda)
    p1_der, p2_der, d_der = divide_y_venceras_1d(derecha)

    # Encontrar la distancia mnima entre las dos mitades
    if d_izq < d_der:
        d_min = d_izq
        p1, p2 = p1_izq, p2_izq
    else:
        d_min = d_der
        p1, p2 = p1_der, p2_der

    # Encontrar la distancia mnima en la franja central
    franja = [x for x in puntos_ordenados if abs(x - punto_medio) < d_min]

    # Comparar puntos en la franja
    for i in range(len(franja)):
        for j in range(i + 1, min(i + 6, len(franja))):  # Solo comparar con los siguientes 5 puntos
            dist = distancia_1d(franja[i], franja[j])
            if dist < d_min:
                d_min = dist
                p1, p2 = franja[i], franja[j]

    return p1, p2, d_min

# Ejecutar la funcin de divide y vencers en 1D
p1, p2, dist = divide_y_venceras_1d(LISTA_1D)

# Mostrar resultados
print("En 1D (Divide y Vencers):")
print(f"Los puntos ms cercanos son {p1} y {p2} con una distancia de {dist}")
```

### Extender El Algoritmo a 2D

```python
# -----Dos puntos mas cercanos----
# Dado un conjunto de puntos se trata de encontrar los dos puntos mas cercanos

# Librerias
import random
import math

# Suponemos las listas de numeros
LISTA_1D = [ random.randrange(1,10000) for x in range(1000)]
LISTA_2D = [(random.randrange(1,10000),random.randrange(1,10000)) for x in range(1000)]

# Funcin para calcular la distancia entre dos puntos en 1D
def distancia_1d(p1, p2):
    return abs(p1 - p2)

# Funcin para calcular la distancia entre dos puntos en 2D
def distancia_2d(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

# Funcin para encontrar los dos puntos ms cercanos en 1D
def puntos_mas_cercanos_1d(puntos):
    n = len(puntos)
    if n < 2:
        return None, None, float('inf')  # No hay suficientes puntos

    punto1, punto2 = puntos[0], puntos[1]
    min_distancia = distancia_1d(punto1, punto2)

    for i in range(n):
        for j in range(i + 1, n):
            dist = distancia_1d(puntos[i], puntos[j])
            if dist < min_distancia:
                min_distancia = dist
                punto1, punto2 = puntos[i], puntos[j]

    return punto1, punto2, min_distancia

# Funcin para encontrar los dos puntos ms cercanos en 2D
def puntos_mas_cercanos_2d(puntos):
    n = len(puntos)
    if n < 2:
        return None, None, float('inf')  # No hay suficientes puntos

    punto1, punto2 = puntos[0], puntos[1]
    min_distancia = distancia_2d(punto1, punto2)

    for i in range(n):
        for j in range(i + 1, n):
            dist = distancia_2d(puntos[i], puntos[j])
            if dist < min_distancia:
                min_distancia = dist
                punto1, punto2 = puntos[i], puntos[j]

    return punto1, punto2, min_distancia

# Ejecutar las funciones
p1_1d, p2_1d, dist_1d = puntos_mas_cercanos_1d(LISTA_1D)
p1_2d, p2_2d, dist_2d = puntos_mas_cercanos_2d(LISTA_2D)

# Mostrar resultados
print("En 1D:")
print(f"Los puntos ms cercanos son {p1_1d} y {p2_1d} con una distancia de {dist_1d}")

print("\nEn 2D:")
print(f"Los puntos ms cercanos son {p1_2d} y {p2_2d} con una distancia de {dist_2d}")


```
